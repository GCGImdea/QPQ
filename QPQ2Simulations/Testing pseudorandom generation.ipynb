```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import chi2_contingency
from scipy.ndimage.interpolation import shift
import seaborn as sns
import scipy.stats as ss
```

```python
rounds = 50000
vals1 = np.random.uniform(0, 1, rounds)
vals2 = np.random.uniform(0, 1, rounds)

sumvals = np.add(vals1, vals2)
meanvals = sumvals/2.0
modvals = sumvals % 1

prodvals = np.multiply(vals1, vals2)
sqrtvals = np.sqrt(prodvals)
modvals2 = sqrtvals % 1

data = np.array([vals1, vals2])

data1 = np.array([vals1, sumvals])
data2 = np.array([vals1, meanvals])
data3 = np.array([vals1, modvals])

data4 = np.array([vals1, prodvals])
data5 = np.array([vals1, sqrtvals])
data6 = np.array([vals1, modvals2])
```

```python
b = plt.plot(data[0], data[1], 'ro')
df = pd.DataFrame({"x":data[0],"y":data[1]})
p = sns.jointplot(data=df,x='x', y='y',kind='kde')
```

```python
chi2, p, dof, expctd = chi2_contingency(data)
print("The test statistic:", chi2)
print("The p-value of the test:", p)
#print("Degrees of freedom:", dof)
#print("The expected frequencies, based on the marginal sums of the table:", expctd)
```
